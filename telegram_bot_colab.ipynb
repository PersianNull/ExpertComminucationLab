```python
import os
from getpass import getpass

print("🔐 لطفاً اطلاعات ربات و حساب کاربری تلگرام خود را وارد کنید:\n")

API_ID = input("API_ID (از my.telegram.org): ")
API_HASH = getpass("API_HASH (نمایش داده نمی‌شود): ")
BOT_TOKEN = getpass("BOT_TOKEN (از BotFather): ")
AUTHORIZED_USERS = input("شناسه عددی کاربر مجاز (مثلاً 123456789): ")

# ساخت فایل .env
env_content = f"""API_ID={API_ID}
API_HASH={API_HASH}
BOT_TOKEN={BOT_TOKEN}
AUTHORIZED_USERS={AUTHORIZED_USERS}
"""

with open(".env", "w") as f:
    f.write(env_content)

print("\n✅ فایل .env با موفقیت ساخته شد.")
print("حالا می‌توانید سلول بعدی را اجرا کنید تا بات فعال شود.")
```
